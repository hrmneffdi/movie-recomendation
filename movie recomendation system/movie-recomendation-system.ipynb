```python
import pandas as pd
import numpy as np
import ast

movie = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
movie.head(1)
```

```python
credit.head(1)
```

```python
data = movie.merge(credit, on='title')
data.head(1)
```

```python
data = data[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
data.head(1)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.iloc[0].genres
```

```python
def extract(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
data['genres'] = data['genres'].apply(extract)
```

```python
data.head(1)
```

```python
data['keywords'] = data['keywords'].apply(extract)
```

```python
def cast(obj):
    batas = 0
    list = []
    for i in ast.literal_eval(obj):
        if batas < 3:
            batas+=1
            list.append(i['name'])
        else:
            break
    return list
```

```python
data['cast'] = data['cast'].apply(cast)
```

```python
data.head(1)
```

```python
def ambil_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return i['name']
```

```python
data['Director'] = data['crew'].apply(ambil_director)
data.drop('crew', axis=1, inplace=True)
data.head(1)
```

```python
data['overview'] = data['overview'].apply(lambda i: i.split())
data.head(1)
```

```python
data['keywords'] = data['keywords'].apply(lambda x : [i.replace(' ','') for i in x])
data.head(1)
```

```python
data['cast'] = data['cast'].apply(lambda x : [i.replace(' ','') for i in x])
data['genres'] = data['genres'].apply(lambda x : [i.replace(' ','') for i in x])
data['Director'] = data['Director'].apply(lambda x : str(x).replace(' ',''))
data.head(1)
```

```python
data['Director'] = data['Director'].apply(lambda x:[x])
data.head(1)
```

```python
data['tags'] = data['overview']+data['genres']+data['keywords']+data['cast']+data['Director']
data
```

```python
film = data[['movie_id', 'title', 'tags']]
film
```

```python
film['tags'] = film['tags'].apply(lambda x : ' '.join(x))
film
```

```python
film['tags'] = film['tags'].apply(lambda x: x.lower())
film.head(1)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return ' '.join(y)
```

```python
film['tags'] = film['tags'].apply(stem)
film.tags[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=4500, stop_words='english')
vector = cv.fit_transform(film['tags']).toarray()
vector, vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity, similarity.shape
```

```python
sorted(list(enumerate(similarity[2])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def system(name_movie):
    index = film[film['title'] == name_movie].index[0]
    recomend = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in recomend:
        print(film.iloc[i[0]].title)
        
system('Spectre')
```
